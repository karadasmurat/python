```python
df = pd.read_csv("data/president_heights.csv")
#print(df)

# get column
heights = df["height(cm)"]
#print(heights)

plt.hist(heights)
plt.title('Height Distribution of US Presidents')
plt.xlabel('height (cm)')
plt.ylabel('number')
```

```python
#pandas.DataFrame.plot
df = pd.DataFrame({'length': [1.5, 0.5, 1.2, 0.9, 3],
                  'width': [0.7, 0.2, 0.15, 0.2, 1.1]},
                  index=['pig', 'rabbit', 'duck', 'chicken', 'horse'])

plot = df.plot(title="DataFrame Plot")
```

```python
data = {'Volkswagen':[10,30,40,50],
        'BMW':[20,40,50,20],
        'Toyota':[30,20,30,40]}
df = pd.DataFrame(data, index=['Small','SUV','Luxury','Pickup'])
df.plot(kind='bar', xlabel='Segment',rot=0, ylabel='in billion USD')
```

```python
data = {'Volkswagen':[10,30,40,50],
        'BMW':[20,40,50,20],
        'Toyota':[30,20,30,40]}
df = pd.DataFrame(data, index=['Small','SUV','Luxury','Pickup'])
df.plot(kind='bar', stacked=True, xlabel='Segment',rot=0, ylabel='in billion USD')
```
